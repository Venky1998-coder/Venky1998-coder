```python
import pyspark
from pyspark.sql.functions import *
from pyspark.sql.functions import rank, col, desc
from pyspark.sql.window import Window
from pyspark.sql import functions as F
from pyspark.sql.types import StructType,StructField,StringType,IntegerType,DoubleType,DateType
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("demo").getOrCreate()
```

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

schema = StructType([
    StructField("customer_id", StringType(), True),
    StructField("customer_name", StringType(), True),
    StructField("signup_date", DateType(), True),
    StructField("age", IntegerType(), True),
    StructField("city", StringType(), True)
])
schema_of_orders = StructType([
    StructField("order_id", IntegerType(), True),
    StructField("customer_id", StringType(), True),
    StructField("order_date", DateType(), True),
    StructField("order_amount", IntegerType(), True),
    StructField("order_status", StringType(), True)
])




df = spark.read.option("header",True).schema(schema).csv("/content/customer.csv")
df.show()
```

```python
df2 = spark.read.option("header",True).schema(schema_of_orders).csv("/content/order.csv")
df2.show()
```

```python
##detecting duplicates
df.groupBy(df.columns).count().filter(F.col("count") > 1).show()
```

```python
df7=df.join(df2,on = "customer_id",how="inner")
df7.show()
```

```python
df2.filter(col("order_amount")>4000).show()
```

```python
df6 = df2.withColumn("telugu",when(col("order_status")=="COMPLETED","vachindi").otherwise("Rale"))
df6.show()
```

```python
df5=df2.dropna(subset=["order_date"]).show()
```

```python
df5.fillna(0,subset=["order_amount"]).show()
```

```python
df8 =df2.fillna(0, subset=["order_amount"])
df8.show()
```

```python
from pyspark.sql.functions import to_date, col

df8 = df2.withColumn(
    "order_date",
    to_date(col("order_date"))
).fillna(
    {"order_date": "2024-10-01"}
).dropna(subset=["order_amount"]).fillna("unknown",subset=["customer_id"])

df8.show()
```

```python
 df2.groupBy(col("order_id")).count().show()
```

```python
df8.dropDuplicates().orderBy(col("order_id")).show()
```

```python
df8.groupBy(col("order_id")).agg(sum("order_amount").alias("total_sales")).show()
```

```python
df8.withColumn("rank",
               rank().over(Window.orderBy(desc(col("order_amount")))

)).show()
```
